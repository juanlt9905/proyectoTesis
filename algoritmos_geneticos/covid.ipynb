```python
df = covid_data1.copy()

#filtro cronologico
df['DateOfConfirmation'] = pd.to_datetime(df['DateOfConfirmation'], errors='coerce')
df = df.dropna(subset=['DateOfConfirmation'])

#Ordenar por fecha
df = df.sort_values(by='DateOfConfirmation', ascending=True)

#obtener los primeros 1495
df_replica = df.head(1495).copy()

print(f"Rango de fechas seleccionado: {df_replica['DateOfConfirmation'].min()} al {df_replica['DateOfConfirmation'].max()}")

```
